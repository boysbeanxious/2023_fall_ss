```python
import os
file_path_list = []
dir_path = "/Users/boysbeanxious/github/samsung/data"

for (root, directories, files) in os.walk(dir_path):
    for file in files:
        if '.xlsx' in file:
            file_path = os.path.join(root, file)
            file_path_list.append(file_path)
```

* # 2019년부터 2023년 대학 계열별 개황 데이터 수집

```python
file_path_list
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter("ignore")
```

```python
tot_df = pd.DataFrame()
for file_name in file_path_list : 
    df = pd.read_excel(file_name,header=[0,1])
    year_col = file_name.split('/')[-1][0:4]
    df['년도'] = year_col
    tot_df = pd.concat([tot_df, df], axis = 0)
```

```python
tot_df.dropna(inplace = True)
```

```python
tot_df = tot_df.set_index([tot_df.iloc[:,0], tot_df.iloc[:,1], tot_df.iloc[:,2]] )
tot_df.index.names = ['대계열', '중계열', '소계열']
```

```python
tot_df.to_excel('./preprocessed.xlsx', encoding = 'cp949', index=True)
```

```python
tot_df.columns
```

* # D1 : 전체 학생 기준 남녀, 지원자 입학자 졸업자 추이 확인

```python
# 대학과정 총계 기준으로 본 재적 학생수, 학과수 지원자, 입학자, 졸업자
D1 = tot_df.loc['대학 과정 총계', ['년도', '재적학생수', '학과수','지원자', '입학자', '졸업자(학사)']].sort_values(by = ['년도'])
```

```python
col_list = D1.columns.to_list()
col_list = [''.join(x) if x[0]!= x[1] else x[0] for x in col_list]
D1.columns = col_list
D1.reset_index(drop=True, inplace=True)
```

```python
D1
```

```python
D1.head()
D1.columns
```

```python
D1[['재적학생수남자비율', '재적학생수여자비율']] = np.round(D1.loc[:, ['재적학생수남자', '재적학생수여자']].div(D1['재적학생수전체'], axis = 0)*100, 2)
```

```python
D1[['년도', '재적학생수남자비율', '재적학생수여자비율']]
```

```python
D1['학과수 증감율'] = np.round(D1[['학과수']].pct_change()*100, 2)
# .pct_change()
```
